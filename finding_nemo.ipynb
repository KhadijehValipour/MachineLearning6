```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 

from knn import KNN
```

```python
nemo = cv2.imread('Nemo.jpg')
nemo = cv2.resize(nemo,(0,0) ,fx=0.25 ,fy=0.25)
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv2.imread('Nemo.jpg')
nemo = cv2.resize(nemo,(0,0) ,fx=0.25 ,fy=0.25)
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)
plt.show()
```

```python
nemo.reshape(-1,3)
```

```python
pixels_list_rgb = nemo.reshape((nemo.shape[0]*nemo.shape[1],3)) / 255.0
R, G ,B = cv2.split(nemo)

fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1, projection='3d')
ax.scatter(R.flatten(), G.flatten(), B.flatten(),facecolors = pixels_list_rgb , marker='.' )
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.show()
```

```python
nemo_hsv = cv2.cvtColor(nemo, cv2.COLOR_BGR2HSV)
H , S ,V = cv2.split(nemo_hsv)
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1 , projection='3d')
ax.scatter(H.flatten() ,S.flatten() , V.flatten() , facecolors = pixels_list_rgb  , marker='.')
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Value')
plt.show()
```

```python
light_orange = (1, 190, 200)
dark_orange = (18, 255, 255)

mask_orange = cv2.inRange(nemo_hsv, light_orange, dark_orange)
result = cv2.bitwise_and(nemo, nemo, mask=mask_orange)

plt.subplot(1, 2, 1)
plt.imshow(mask_orange, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
