```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 

from knn import KNN
```

```python
nemo = cv2.imread('Nemo.jpg')
nemo = cv2.resize(nemo,(0,0) ,fx=0.25 ,fy=0.25)
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv2.imread('Nemo.jpg')
nemo = cv2.resize(nemo,(0,0) ,fx=0.25 ,fy=0.25)
hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv2.imread('Nemo.jpg')
nemo = cv2.resize(nemo,(0,0) ,fx=0.25 ,fy=0.25)
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)
plt.show()
```

```python
nemo.reshape(-1,3)
```

```python
pixels_list_rgb = nemo.reshape(-1,3)

pixels_list_rgb.shape

fig = plt.figure(figsize=(15,13))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pixels_list_rgb[:,0] , pixels_list_rgb[:,1] , pixels_list_rgb[:,2] , c=pixels_list_rgb/255 )
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.show()
```

```python
nemo_hsv = cv2.cvtColor(nemo, cv2.COLOR_BGR2HSV)
#plt.imshow(nemo_hsv)
```

```python
nemo_hsv.reshape(-1,3)
```

```python
pixels_list_hsv = nemo_hsv.reshape(-1,3)
fig = plt.figure(figsize=(15,13))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(pixels_list_hsv[:,0],pixels_list_hsv[:,1],pixels_list_hsv[:,2], c= pixels_list_rgb/255)
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Value')
plt.show()
```

```python
light_orange = (1, 190, 200)
dark_orange = (18, 255, 255)

mask_orange = cv2.inRange(hsv_nemo, light_orange, dark_orange)
result_orange = cv2.bitwise_and(nemo, nemo, mask=mask_orange)

plt.subplot(1, 2, 1)
plt.imshow(mask_orange, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result_orange)
plt.show()
```
